```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team two
team_two_file_path = os.path.join('../','Resources', 'team-2.csv')
```

```python
# Read the CSV file into a DataFrame
team_two_df = pd.read_csv(team_two_file_path)
team_two_df
```

```python
# Melt the DataFrame
team_two_df.melt()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team_two_df.melt(id_vars= 'runner_name')
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team_two_df.melt(id_vars= 'coach_name')
```

```python
# Renaming the columns
team_two_df.melt(id_vars="runner_name", var_name="run", value_name="distance")
```

```python
# Remove coach name from DataFrame
del team_two_df['coach_name']
team_two_df
```

```python
# Show final result
final_df= team_two_df.melt(id_vars="runner_name", var_name="lap", value_name="distance")
```

```python
# Save results to a file
final_df= team_two_df.to_csv("melted_running_df.csv", sep=',')
```
