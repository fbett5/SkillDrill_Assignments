```python
# Import dependencies
import pandas as pd
```

```python

# Read file into DataFrame
ratings_and_sent_review_df = pd.read_csv('../Resources/ratings_and_sentiments.csv')
ratings_and_sent_review_df.head(5)
```

```python
# Find all reviews with a number rating greater than or equal to 4.
ratings_df = ratings_and_sent_review_df[ratings_and_sent_review_df['num_rating']>=4]
ratings_df.head()
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
overall_df = ratings_and_sent_review_df[ratings_and_sent_review_df['overall_sent']>=4]
overall_df.head()
```

```python
# Return the count of the above query.
len(ratings_and_sent_review_df[(ratings_and_sent_review_df['overall_sent'] >= 4) & (ratings_and_sent_review_df['num_rating'] >=4)])
```

```python
# Change the name of the review_text column to review
ratings_and_sent_review_df.rename(columns= {'review_text':'review'}, errors="raise").head()
```

```python
# Change the name of the bool_HIGH column to bool_high
ratings_and_sent_review_df.rename(columns= {'bool_HIGH':'bool_high'}, errors = 'raise').head()
```

```python
# Save your changes to the file as a pipe delimited file.
ratings_and_sent_review_df.to_csv('ratings_and_sent_review_df.txt', sep = '|')
```
